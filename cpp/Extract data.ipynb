```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
%config InlineBackend.rc = {'figure.figsize': (14., 8.)}
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
class Result(object):
    def __init__(self, eta, B, res_as_np_array=None):
        if res_as_np_array:
            self.data = res_as_np_array
        self.eta = eta
        self.B = B
    def plot(self, mask = 0):
        import matplotlib.pyplot as plt
        if not mask == None:
            self.data_masked = np.ma.masked_less_equal(self.data, mask)
        else:
            self.data_masked = self.data
        plt.matshow(self.data_masked, extent=(0, 31, -5.0, 0), interpolation='none', aspect='auto', vmin=0)
        plt.xlabel("$\Omega / \Omega_\odot$")
        plt.ylabel("$\partial \log \Omega / \partial \log r$")
        plt.title(r"FGM with $\eta = %g$ and $B_{\theta} = %g$ -- in CGS " % (self.eta, self.B))
        plt.colorbar()
        plt.grid()
        #plt.gca().invert_yaxis()
        plt.show()
    def load_data(self, path, first_line=0):
        import numpy as np
        # skip the first_line
        for _ in range(first_line):
            f.readline()
        # read the data
        with open(path, 'r') as f:
            self.data = np.array([ [ float(e)/2.7e-6 for e in line.split() ] for line in f.readlines() ])
            assert(len(self.data.shape) == 2)
        
```

```python
res1 = Result(496, 400)
res1.load_data("results/b_400.log")
res1.plot(mask = 1)

res2 = Result(496, 40)
res2.load_data("results/b_40.log")
res2.plot(mask = 1)

plt.title("Normalized difference between $B=40$ and $B=400$ (log)")
plt.imshow(np.log10(np.abs(2*(res1.data - res2.data)/(res1.data + res2.data))),
           extent=(0, 31, -5.0, 0), interpolation='none', aspect='auto')
plt.colorbar()

res3 = Result(496, 4)
res3.load_data("results/b_4.log")
res3.plot(mask = 1)

plt.title("Normalized difference between $B=400$ and $B=4$ (log)")
plt.imshow(np.log10(np.abs(2*(res1.data - res3.data)/(res1.data + res3.data))),
           extent=(0, 31, -5.0, 0), interpolation='none', aspect='auto')
plt.colorbar()

res3 = Result(496, 0)
res3.load_data("results/b_0.log")
res3.plot(mask = 1)
```
