```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
f = open("FGM")
FGM = np.array([ [ float(e) for e in line.split() ] for line in f.readlines() ])
f = open("FGM_kR")
FGM_kR = np.array([ [ float(e) for e in line.split() ] for line in f.readlines() ])
f = open("FGM_kZ")
FGM_kZ = np.array([ [ float(e) for e in line.split() ] for line in f.readlines() ])
```

```python
plt.matshow(FGM/2.7e-6, extent=(0, 31, -5.0, 0), interpolation='none', aspect='auto', vmin=0)
plt.colorbar()
```

```python
plt.matshow(FGM_kR, extent=(0, 31, -5.0, 0), interpolation='none', aspect='auto', vmin=0)
plt.colorbar()
```

```python
plt.matshow(FGM_kZ, extent=(0, 31, -5.0, 0), interpolation='none', aspect='auto')
plt.colorbar()
```

```python
plt.figure(figsize=(12,12))
plt.imshow(FGM/(FGM_kR*FGM_kR), extent=(0, 31, -5.0, 0), interpolation="none", aspect="auto")
plt.colorbar()
```
